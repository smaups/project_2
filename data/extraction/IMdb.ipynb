```python
import pandas as pd
```

```python
imdb_data = "Data/imdb.csv"
```

```python
##had to use latin-1 encoding for it to be able to read numbers right
imdb_df = pd.read_csv(imdb_data, encoding='latin-1')
```

```python
##converting the imdbID column to string
imdb_df['imdbID'] = imdb_df['imdbID'].astype(str)
```

```python
imdb_df
```

```python
df1 = pd.DataFrame()

```

```python
for index, row in imdb_df.iterrows():
    url = "https://www.imdb.com/title/" + row["imdbID"] + "/fullcredits/"
    try:
        tables = pd.read_html(url)
        df = tables[2]
        df = df.dropna(how = 'all')
        del df[0]
        del df[2]
        df = df.head(10)
        df["imdb"] = row["imdbID"]
        df1 = df.append(df1) 
    except:
        print("Missing values for " + row["imdbID"])

```

```python
df1
```

```python
df1.describe()
```

```python
list_imdbID = list(df1["imdbID"].unique())
list_imdbID

    
```

```python
imdb_dict = {}
for movie in list_imdbID:
    temp_df = df1.loc[df1['imdbID'] == movie]
    imdb_dict[movie] = list(temp_df[3])
print(imdb_dict['tt6306064'])
    
```

```python
imdb_dict_final = {}
for key in imdb_dict:
    imdb_dict1 = imdb_dict[key]
    x = str(imdb_dict1[0])
    for i in range(1, len(imdb_dict1)):
        x = x + ', ' + str(imdb_dict1[i])
    #print(x)
    imdb_dict_final[key] = x
imdb_dict_final
```

```python
final_df = pd.DataFrame(list(imdb_dict_final.items()))
final_df
```

```python
final_df.to_csv("final_character_names.csv", index=False, header=True)
```
